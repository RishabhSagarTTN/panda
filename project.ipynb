```python
df.drop(['director'],axis=1,inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.groupby(['country']).value_counts().sort_values().loc['United States']
```

```python
import matplotlib.pyplot as plt
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar', title='Top 10 Content Producing Countries')
plt.ylabel('Number of Titles')
plt.show()
```

```python
df['listed_in'].str.split(', ').explode().value_counts().head(10).plot(kind='barh', title='Top 10 Genres')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()
plt.show()

```

```python

# Distribution of Ratings
df['rating'].value_counts().plot(kind='bar', title='Distribution of Content Ratings')
plt.ylabel('Number of Titles')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],format='mixed')
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='line', title='Content Added Over the Years')
plt.ylabel('Number of Titles')
plt.show()
```
